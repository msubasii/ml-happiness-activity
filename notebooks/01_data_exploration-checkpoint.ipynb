# Physical Activity & Happiness — Data Exploration

This notebook loads, merges, and cleans three datasets used in this project:
- **World Happiness Report 2015** — happiness scores and contributing factors (GDP, health, freedom, etc.)
- **World Happiness Index 2013–2023** — multi-year happiness scores for trend analysis
- **WHO Physical Activity Country Profiles 2022** — physical inactivity rates by age group and gender

The goal is to produce a single clean dataset (`merged_clean.csv`) ready for EDA and modeling.

## 1. Load Datasets

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Load the 2015 happiness report — contains detailed feature columns (GDP, family, health, etc.)
happiness_2015 = pd.read_csv('../data/world_happiness.csv')

# Load the multi-year happiness index (2013–2023) — used to compute per-country averages
happiness_allyears = pd.read_csv('../data/world_happiness_allyears.csv')

# Load WHO physical activity profiles — contains inactivity rates by age group and gender
who = pd.read_csv('../data/who_physical_activity.csv')

print("Happiness 2015 shape:", happiness_2015.shape)
print("Happiness all years shape:", happiness_allyears.shape)
print("WHO shape:", who.shape)
print()
print("=== All years columns ===")
print(happiness_allyears.columns.tolist())
print()
print("=== All years sample ===")
print(happiness_allyears.head(3))
```

## 2. Compute Average Happiness Score (2013–2023)

Since the WHO dataset is a single 2022 snapshot (not yearly), we cannot do a year-by-year merge.
Instead, we compute each country's average happiness score across all available years.
This gives a more stable, long-term measure of happiness rather than relying on a single year.

```python
# Group by country and take the mean of the happiness index across all years
happiness_avg = happiness_allyears.groupby('Country')['Index'].mean().reset_index()
happiness_avg.columns = ['Country', 'Happiness Score Avg']

print("Countries with avg score:", len(happiness_avg))
print(happiness_avg.head(5))
```

## 3. Merge Datasets

We merge all three datasets on the `Country` column using an inner join,
keeping only countries that appear in all three sources.
This reduces the dataset from 158/167/194 countries to 131 matched countries.

> **TODO:** Fix country name mismatches (e.g. 'Turkiye' vs 'Turkey') to recover the ~27 dropped countries.

```python
# First merge: 2015 happiness features + multi-year average score
df = pd.merge(happiness_2015, happiness_avg, on='Country', how='inner')

# Second merge: add WHO physical activity data
df = pd.merge(df, who, on='Country', how='inner')

print("Final merged shape:", df.shape)
print()
print("Columns:", df.columns.tolist())
```

## 4. Data Cleaning

The WHO dataset stores inactivity rates as strings with a `%` symbol (e.g. `'42.0%'`).
We strip the symbol and convert to numeric floats.
Missing values are filled with the column mean — a simple imputation strategy suitable for this dataset size.

We also create two summary columns:
- `Avg Inactivity Adults` — average of male and female inactivity for ages 18+
- `Avg Inactivity Adolescents` — average of male and female inactivity for ages 11–17

```python
who_cols = [
    'Physical Inactivity, Adolescents Aged 11-17 (%), Male',
    'Physical Inactivity, Adolescents Aged 11-17 (%), Female',
    'Physical Inactivity, Adults Aged 18+ (%), Male',
    'Physical Inactivity, Adults Aged 18+ (%), Female',
    'Physical Inactivity, Adults Aged 70+ (%), Male',
    'Physical Inactivity, Adults Aged 70+ (%), Female'
]

# Remove '%' sign and convert to float
for col in who_cols:
    df[col] = df[col].astype(str).str.replace('%', '').str.strip()
    df[col] = pd.to_numeric(df[col], errors='coerce')

# Fill remaining missing values with column mean
df[who_cols] = df[who_cols].fillna(df[who_cols].mean())

# Create average inactivity columns for adults and adolescents
df['Avg Inactivity Adults'] = df[[
    'Physical Inactivity, Adults Aged 18+ (%), Male',
    'Physical Inactivity, Adults Aged 18+ (%), Female'
]].mean(axis=1)

df['Avg Inactivity Adolescents'] = df[[
    'Physical Inactivity, Adolescents Aged 11-17 (%), Male',
    'Physical Inactivity, Adolescents Aged 11-17 (%), Female'
]].mean(axis=1)

print("Missing values after cleaning:")
print(df.isnull().sum()[df.isnull().sum() > 0])
print()
print(df[['Country', 'Happiness Score', 'Happiness Score Avg', 'Avg Inactivity Adults']].head(5))
```

## 5. Fix Remaining Missing Values

The `GDP per capita (INT$)` column from the WHO dataset has 2 missing values.
We convert it to numeric and fill with the column mean.

```python
# Convert GDP column to numeric (some values may have formatting issues)
df['GDP per capita (INT$)'] = pd.to_numeric(
    df['GDP per capita (INT$)'], errors='coerce'
)

# Fill 2 missing GDP values with the column mean
df['GDP per capita (INT$)'] = df['GDP per capita (INT$)'].fillna(
    df['GDP per capita (INT$)'].mean()
)

# Save the final cleaned dataset for use in modeling notebooks
df.to_csv('../data/merged_clean.csv', index=False)
print("Saved merged_clean.csv — Shape:", df.shape)
```

```python
# Fix population column - remove commas and convert to numeric
df['Total Population (2022)'] = df['Total Population (2022)'].astype(str).str.replace(',', '').str.strip()
df['Total Population (2022)'] = pd.to_numeric(df['Total Population (2022)'], errors='coerce')

# Fix percentage columns - remove % sign and convert to numeric
pct_cols = ['Population Aged < 20 (%)', 'Population Aged > 60 (%)', 'Deaths Due to NCD']
for col in pct_cols:
    df[col] = df[col].astype(str).str.replace('%', '').str.strip()
    df[col] = pd.to_numeric(df[col], errors='coerce')

# Overwrite the saved file with fixed version
df.to_csv('../data/merged_clean.csv', index=False)

print("Fixed! Sample:")
print(df[['Country', 'Total Population (2022)', 'Population Aged < 20 (%)', 'Population Aged > 60 (%)']].head(5))
print()
print("Dtypes:")
print(df[['Total Population (2022)', 'Population Aged < 20 (%)', 'Population Aged > 60 (%)']].dtypes)
```
